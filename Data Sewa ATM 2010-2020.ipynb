```python
import pandas as pd
import numpy as np
```

```python
#Membaca data dari baris ke-3
df = df = pd.read_excel ('data-set.xlsx', sheet_name='SEWAATM', header=4, skiprows=2)
```

```python
df.head()
```

```python
#Merubah nama kolom yang sesuai
df = df.rename({'Unnamed: 1': 'Tanggal Request', 'Unnamed: 2': 'Nama Lokasi', 'Unnamed: 3': 'Kota', 'Unnamed: 4': 'Masa Sewa', 'Unnamed: 5': 'Delivery', 'Unnamed: 6': 'Type', 'Unnamed: 7': 'Jatuh Tempo' }, axis=1)
```

```python
df.head()
```

```python
#Drop kolom nomor
df = df.drop('Unnamed: 0', 1)
```

```python
#Menggunakan kolom yang akan digunakan
df = df[['Tanggal Request', 'Nama Lokasi', 'Kota', 'Masa Sewa', 'Delivery', 'Type', 'Jatuh Tempo']]
```

```python
df.head()
```

```python
#Cek unique value dari kolom Kota
df['Kota'].unique()
```

```python
#Cek apakah ada kata cancel atau off pada kolom
def func(a):
    if 'cancel' in a.lower():
        return 'cancel'
    elif 'off' in a.lower():
        return 'off'
    else:
        return 'aktif'

#Membuat kolom Status yang berisi fungsi diatas
df['Status'] = df.Kota.apply(lambda x: func(x))
```

```python
df.head()
```

```python
#Cek uniquq value kolom Status
df['Status'].unique()
```

```python
#Cek unique value dari kolom Masa Sewa
df['Masa Sewa'].unique()
```

```python
#Cuma ada satu row yang berisi data 1 BULAN, maka diputuskan untuk didrop
df.drop(index=df[df['Masa Sewa'] == '1 BULAN'].index, inplace=True)
```

```python
df['Masa Sewa'].unique()
```

```python
#Cek unique value dari kolom Delivery
df['Delivery'].unique()
```

```python
#Merubah kolom Delivery menjadi bertipe data datetime dan menghapus data yang bukan bertipe datetime
df['Delivery'] = pd.to_datetime(df['Delivery'], errors='coerce')
```

```python
df['Delivery'].unique()
```

```python
df.head(5)
```

```python
#Cek unique value dari kolom Type
df['Type'].unique()
```

```python
#Menghapus value yang tidak penting
df['Type'].replace('-', np.nan, inplace = True)
df['Type'].replace('cancel', np.nan, inplace = True)
df['Type'].replace('Cancel', np.nan, inplace = True)
```

```python
df['Type'].unique()
```

```python
df.head(5)
```

```python
df.info()
```

```python
#Merubah tipe data kolom Tanggal Request menjadi datetime dan menghapus row yang memiliki tipe data bukan datetime
df['Tanggal Request'] = pd.to_datetime(df['Tanggal Request'], errors='coerce')
```

```python
#Merubah tipe data kolom Masa Sewa menjadi float64
df[['Masa Sewa']] = df[['Masa Sewa']].astype('float')
```

```python
#Cek unique value kolom Jatuh Tempo
df['Jatuh Tempo'].unique()
```

```python
#Cek data dengan value Agt-24
df.loc[df['Jatuh Tempo'] == 'Agt-24']
```

```python
#Merubah format value Agt-24 agar mudah diganti ke bentuk datetime
df['Jatuh Tempo'].replace('Agt-24', '08/01/2024', inplace = True)
```

```python
#Merubah tipe data kolom Jatuh Tempo menjadi datetime
df['Jatuh Tempo'] = pd.to_datetime(df['Jatuh Tempo'], errors='coerce')
```

```python
df['Jatuh Tempo'].unique()
```

```python
#Cek unique value kolom Kota
df['Kota'].unique()
```

```python
#Mengesktrak kata pertama dari tiap value pada kolom Nama Kota
df['Nama Kota'] = df.Kota.str.extract(r'^\W*([\w-]+)', expand=True)
df['Nama Kota'].unique()
```

```python
#Lowercase value pada kolom Nama Kota
df['Nama Kota'] = df['Nama Kota'].str.lower()
```

```python
df['Nama Kota'].unique()
```

```python
#Memperbaiki kesalahan penamaan kota yang belum sesuai
df['Nama Kota'] = df['Nama Kota'].str.replace('jkt-hub','jakarta')
df['Nama Kota'] = df['Nama Kota'].str.replace('solo-cancel','solo')
df['Nama Kota'] = df['Nama Kota'].str.replace('bandar','bandar lampung')
df['Nama Kota'] = df['Nama Kota'].str.replace('cirebon-rpc','cirebon')
df['Nama Kota'] = df['Nama Kota'].str.replace('cikarang-off','cikarang')
df['Nama Kota'] = df['Nama Kota'].str.replace('banda','banda aceh')
df['Nama Kota'] = df['Nama Kota'].str.replace('tanjung','tanjung pinang')
df['Nama Kota'] = df['Nama Kota'].str.replace('jkt-dpti','jakarta')
df['Nama Kota'] = df['Nama Kota'].str.replace('cirebon-off','cirebon')
df['Nama Kota'] = df['Nama Kota'].str.replace('manado-cancel','manado')
```

```python
df['Nama Kota'].unique()
```

```python
#Memperbaiki kesalahan penamaan kota yang belum sesuai
df['Nama Kota'] = df['Nama Kota'].str.replace('banda acehr lampung','bandar lampung')
```

```python
df['Nama Kota'].unique()
```

```python
#Title case kolom Nama Kota
df['Nama Kota'] = df['Nama Kota'].str.title()
```

```python
df['Nama Kota'].unique()
```

```python
#Menyusun ulang urutan kolom untuk finalisasi
df = df[['Tanggal Request', 'Nama Lokasi','Kota', 'Nama Kota', 'Masa Sewa', 'Delivery', 'Type', 'Jatuh Tempo', 'Status']]
df.head()
```

```python
df.to_excel('Sewa ATM.xlsx')
```
